## Exercice 2

````python
def matriceDiag(n):
    B = np.zeros((n,n))
    B[0][0] = 2
    for i in range(n):
        for j in range(0,n):
            if i == j:
                B[i][j] = 2
            elif i == j+1 or j == i+1:
                B[i][j] = -1
            else :
                B[i][j] = 0
    return B



b5 = matriceDiag(5)
badis = matriceDiag(10)
b50 = matriceDiag(50)

valeurs_propresB, vecteurs_propresB = np.linalg.eig(b5)

print("B avec n=5:\n",b5)
print("Valeurs propres :\n", valeurs_propresB,"\n")
print("Vecteurs propres :\n", vecteurs_propresB,"\n")

valeurs_propresB, vecteurs_propresB = np.linalg.eig(badis)

print("B avec n=10:\n",badis)
print("Valeurs propres :\n", valeurs_propresB,"\n")

print("Vecteurs propres :\n", vecteurs_propresB,"\n")
valeurs_propresB, vecteurs_propresB = np.linalg.eig(b50)

print("B avec n=50:\n",b50)
print("Valeurs propres :\n", valeurs_propresB,"\n")
print("Vecteurs propres :\n", vecteurs_propresB,"\n")
````

### TD0

Reprendre l'exercice 5 du TD0 et construire la représentation graphique de $x\mapsto \ln(1+x)$ et de son polynôme osculatoire $p$ de degré $999$ en $x=0$. Quelle est l'erreur commise pour $x=1$ lorsqu'on remplace la fonction $x\mapsto \ln(1+x)$ par $x\mapsto p(x)$ ?

````python
x = np.linspace(0.01, 0.99, 100)
y=[]

for i in range(len(x)):
    y.append(m.log(1+x[i]))

y2 = []
N = 10
for xi in x:
    taylor_sum = 0
    for n in range(1, N + 1):
        taylor_sum += ((-1) ** (n + 1)) * (xi ** n) / n
    y2.append(taylor_sum)

y3 = []
N = 100000
for xi in x:
    taylor_sum = 0
    for n in range(1, N + 1):
        taylor_sum += ((-1) ** (n + 1)) * (xi ** n) / n
    y3.append(taylor_sum)



plt.figure(figsize=(10, 6))  
plt.plot(x, y, color='purple', label="$f:x\\mapsto\\ln(1+x)$")
plt.plot(x, y2, color='green', label="Poly pour N=10")
plt.plot(x, y3, color='c', label="Poly pour N=999")


plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("")
plt.legend()
plt.show()

````

### Implémentez les algorithmes de Horner et Naif et tester les temps de calcul pour des polynômes aléatoires en prenant des valeurs de plus en plus grandes de n

````python
def algoNaif(p,x):
    n = len(p)
    S = p[0]
    for i in range(1,n):
        t = 1
        for j in range(1, i+1):
            t*=x
        S+=p[i]*t
    return S


def horner(p, x):
    n = len(p)
    val = p[n-1]
    for i in range(n-2, -1, -1):
        val=val*x+p[i]
    return val
    

print(algoNaif([1,1,1],3))
print(horner([1,1,1],3))
````

````python
import random
import time

def createRandomSortedList(num, start, end): 
    arr = [] 
    tmp = random.randint(start, end) 
      
    for x in range(num): 
          
        while tmp in arr: 
            tmp = random.randint(start, end) 
              
        arr.append(tmp) 
          
    arr.sort() 
      
    return arr 


def createRandomUnSortedList(num, start, end): 
    arr = [] 
    tmp = random.randint(start, end) 
      
    for x in range(num): 
          
        while tmp in arr: 
            tmp = random.randint(start, end) 
              
        arr.append(tmp) 
      
    return arr 
````

````python
x = []
y1 = []
y2 = []

n = 1
repeat = 1000


for i in range(30):
    array = createRandomUnSortedList(n, 0, 2000)
    random_value = random.randint(0, 1000)
    
    for _ in range(repeat):
        start_time = time.time()
        algoNaif(array,random_value)
    y1.append((time.time() - start_time)/repeat)

    for __ in range(repeat):
        start_time = time.time()
        horner(array,random_value)
    y2.append((time.time() - start_time)/repeat)
    
    

    n+=10
    x.append(n)

plt.figure(figsize=(10, 6))  
plt.plot(x, y1, color='r', label="Algo naif")
plt.plot(x, y2, color='c', label="Horner")
plt.xlabel("taille polynôme")
plt.ylabel("temps en sec")
plt.title("Complexité : Algo naif VS Horner")
plt.legend()
plt.show()
````
